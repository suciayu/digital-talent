### One to many relationship

```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def RUN(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#1. Report the account representative for each customer.
RUN("""
SELECT CONCAT(E.FIRSTNAME, ' ' , E.LASTNAME) AS REPNAME, C.CUSTOMERNAME 
FROM EMPLOYEES E INNER JOIN CUSTOMERS C 
WHERE E.EMPLOYEENUMBER = C.SALESREPEMPLOYEENUMBER
""")
```

```python
#2. Report total payments for Atelier graphique
RUN("""
SELECT SUM(AMOUNT) AS TOTALPAYMENT
FROM CUSTOMERS INNER JOIN PAYMENTS
USING(CUSTOMERNUMBER) 
WHERE CUSTOMERNAME='Atelier Graphique'
""")
```

```python
#3 Report the total payments by date
RUN("""
SELECT PAYMENTDATE, SUM(AMOUNT) AS TOTALPAYMENTS
FROM PAYMENTS
GROUP BY PAYMENTDATE
""")
```

```python
#4 Report the products that have not been sold
RUN("""
SELECT *
FROM PRODUCTS LEFT JOIN ORDERDETAILS
USING(PRODUCTCODE)
WHERE QUANTITYORDERED IS NULL
""")
```

```python
#5. List the amount paid by each customer
RUN("""
SELECT C.CUSTOMERNUMBER, C.CUSTOMERNAME, SUM(AMOUNT) AS AMOUNTPAID
FROM CUSTOMERS C LEFT JOIN PAYMENTS P
USING(CUSTOMERNUMBER)
GROUP BY CUSTOMERNUMBER
""")
```

```python
#6. How many orders have been placed by Herkku Gifts?
RUN("""
SELECT COUNT(*) AS TOTALORDER
FROM CUSTOMERS C INNER JOIN ORDERS O
USING(CUSTOMERNUMBER)
WHERE C.CUSTOMERNAME='Herkku Gifts'
""")
```

```python
#7. Who are the employees in Boston?
RUN("""
SELECT CONCAT(FIRSTNAME, ' ', LASTNAME) AS NAME, EMAIL, JOBTITLE, PHONE, ADDRESSLINE1, ADDRESSLINE2, CITY
FROM EMPLOYEES E INNER JOIN OFFICES O
USING(OFFICECODE)
WHERE CITY='BOSTON'
""")
```

```python
#8. Report those payments greater than \$100,000. Sort the report so the customer who made the highest payment appears first.
RUN("""
SELECT CUSTOMERNAME, AMOUNT
FROM CUSTOMERS C INNER JOIN PAYMENTS P
USING(CUSTOMERNUMBER)
WHERE AMOUNT > 100000
ORDER BY AMOUNT
""")
```

```python
#9. List the value of 'On Hold' orders.
RUN("""
SELECT O.ORDERNUMBER , SUM(QUANTITYORDERED*PRICEEACH) AMOUNT
FROM ORDERS O INNER JOIN ORDERDETAILS D
USING(ORDERNUMBER)
WHERE STATUS = 'ON HOLD'
GROUP BY ORDERNUMBER
""")
```

```python
#10. Report the number of orders 'On Hold' for each customer.
RUN("""
SELECT C.CUSTOMERNAME, O.ORDERNUMBER
FROM CUSTOMERS C INNER JOIN ORDERS O
USING(CUSTOMERNUMBER)
WHERE STATUS='ON HOLD'
""")
```

### MANY TO MANY RELATIONSHIP

```python
# 1. List products sold by order date.
RUN("""
SELECT ORDERDATE, SUM(QUANTITYORDERED) AS PRODUCTSOLD
FROM ORDERS O JOIN ORDERDETAILS
USING(ORDERNUMBER)
GROUP BY ORDERDATE
""")
```

```python
# 2. List the order dates in descending order for orders for the 1940 Ford Pickup Truck
RUN("""
SELECT ORDERDATE, PRODUCTNAME
FROM ORDERS O JOIN ORDERDETAILS D JOIN PRODUCTS P 
ON O.ORDERNUMBER = D.ORDERNUMBER AND P.PRODUCTCODE=D.PRODUCTCODE
WHERE PRODUCTNAME = '1940 Ford Pickup Truck'
GROUP BY ORDERDATE
ORDER BY ORDERDATE DESC
""")
```

```python
#3. List the names of customers and their corresponding order number where a particular order from that customer has a value greater than $25,000?
RUN("""
SELECT C.CUSTOMERNAME, SUM((PRICEEACH*QUANTITYORDERED)) AS AMOUNT
FROM CUSTOMERS C RIGHT JOIN ORDERS O USING(CUSTOMERNUMBER) JOIN ORDERDETAILS D USING(ORDERNUMBER)
GROUP BY ORDERNUMBER
HAVING AMOUNT > 25000
""")
```

```python
# 4. Are there any products that appear on all orders? NO
RUN("""
SELECT PRODUCTCODE, COUNT(*) AS APPEARANCE
FROM PRODUCTS P JOIN ORDERDETAILS D
USING(PRODUCTCODE)
GROUP BY PRODUCTCODE
HAVING APPEARANCE = (SELECT COUNT(*) FROM ORDERS)
""")
```

```python
#5. List the names of products sold at less than 80% of the MSRP.
RUN("""
SELECT DISTINCT PRODUCTNAME, MSRP, PRICEEACH, ((PRICEEACH/MSRP)*100) AS PERCENTAGE
FROM PRODUCTS P JOIN ORDERDETAILS D
USING(PRODUCTCODE)
HAVING PERCENTAGE < 80
""")
```

```python
#6. Reports those products that have been sold with a markup of 100% or more (i.e.,  the priceEach is at least twice the buyPrice)
RUN("""
SELECT DISTINCT PRODUCTNAME
FROM PRODUCTS P JOIN ORDERDETAILS D
USING(PRODUCTCODE)
WHERE PRICEEACH >= 2*BUYPRICE
""")
```

```python
#7. List the products ordered on a Monday (0)
RUN("""
SELECT DISTINCT PRODUCTNAME, WEEKDAY(ORDERDATE) AS DAY
FROM ORDERS O JOIN ORDERDETAILS D USING(ORDERNUMBER) JOIN PRODUCTS USING(PRODUCTCODE)
HAVING DAY = 0
""")
```

```python
#8. What is the quantity on hand for products listed on 'On Hold' orders?
RUN("""
SELECT PRODUCTNAME, QUANTITYINSTOCK
FROM ORDERS O JOIN ORDERDETAILS USING(ORDERNUMBER) JOIN PRODUCTS USING(PRODUCTCODE)
WHERE STATUS = 'ON HOLD'
""")
```
