```python
import pymysql
import pandas as pd

host = '127.0.0.1'
port = "3306"
user = 'root'
password = ''
database = 'classicmodels'

conn = pymysql.connect(
    host=host,
    port=int(port),
    user=user,
    passwd=password,
    db=database,
    charset='utf8mb4')

def run(sql):
    df = pd.read_sql_query(sql,conn)
    return df
```

```python
#Prepare a list of offices sorted by country, state, city.
run("SELECT * FROM OFFICES ORDER BY COUNTRY, STATE, CITY")
```

```python
#How many employees are there in the company?
run("SELECT COUNT(*) FROM EMPLOYEES AS TOTAL_EMPLOYEE")
```

```python
#What is the total of payments received?
run("SELECT SUM(AMOUNT) FROM PAYMENTS AS TOTAL_PAYMENT")
```

```python
#List the product lines that contain 'Cars'.
run("SELECT * FROM PRODUCTLINES WHERE PRODUCTLINE LIKE '%Car%'")
```

```python
#Report total payments for October 28, 2004
run("SELECT SUM(AMOUNT) FROM PAYMENTS WHERE PAYMENTDATE='2004-10-28'")
```

```python
#Report those payments greater than \$100,000.
run("SELECT * FROM PAYMENTS WHERE AMOUNT>100000")
```

```python
#List the products in each product line.
run("SELECT PRODUCTLINE, PRODUCTNAME FROM PRODUCTS ORDER BY PRODUCTLINE")
```

```python
#How many products in each product line?
run("SELECT PRODUCTLINE, COUNT(PRODUCTCODE) FROM PRODUCTS GROUP BY PRODUCTLINE")
```

```python
#What is the minimum payment received?
run("SELECT MIN(AMOUNT) FROM PAYMENTS")
```

```python
run("SELECT * FROM PRODUCTS")
```

```python
#List all payments greater than twice the average payment.
run("SELECT * FROM PAYMENTS WHERE AMOUNT>(SELECT AVG(AMOUNT)*2 FROM PAYMENTS)")
```

```python
#What is the average percentage markup of the MSRP on buyPrice?
run("SELECT AVG((MSRP-BUYPRICE)/BUYPRICE) FROM PRODUCTS")
```

```python
#How many distinct products does ClassicModels sell?
run("SELECT COUNT(*) FROM PRODUCTS WHERE PRODUCTLINE LIKE 'Classic%'")
```

```python
#Report the name and city of customers who don't have sales representatives?
run("SELECT customername, city FROM CUSTOMERS where salesRepEmployeeNumber is null")
```

```python
#What are the names of executives with VP or Manager in their title? 
#Use the CONCAT function to combine the employee's first name and last name into a single field for reporting.
run("SELECT CONCAT(firstName,' ',lastName) AS fullName, JOBTITLE FROM EMPLOYEES WHERE JOBTITLE LIKE 'VP%' OR JOBTITLE LIKE '%Manager%'")
```

```python
#Which orders have a value greater than $5,000?
run("SELECT ORDERNUMBER, PRODUCTCODE, (PRICEEACH*QUANTITYORDERED) AS PRICE FROM ORDERDETAILS HAVING PRICE>5000")
```
